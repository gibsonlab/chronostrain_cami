```python
CHRONOSTRAIN_DB_DIR = Path("/mnt/e/CAMI_strain_madness/inference/databases/ecoli")
GOLD_STANDARD_INDEX = Path("/mnt/e/CAMI_strain_madness/gold_standard_genomes") / "index.tsv"
MARKER_SEED_INDEX = Path().resolve().parent.parent.parent / "data" / "seeds" / "ecoli" / "marker_seed_index.tsv"

""" Reference Genome/Isolate BLAST database """
BLAST_DB_DIR = CHRONOSTRAIN_DB_DIR / "blast_db"  # The location of the Blast DB that you wish to create using RefSeq indices (they will be downloaded by this notebook).
BLAST_DB_NAME = "CAMI_Strain_Madness_Gold"  # Blast DB to create.

""" chronostrain-specific settings """
NUM_CORES = 8  # number of cores to use (e.g. for blastn)
MARKER_MIN_PCT_IDTY = 75  # accept BLAST hits as markers above this threshold
CHRONOSTRAIN_TARGET_JSON = CHRONOSTRAIN_DB_DIR / "database.json"

""" Some checks. """
assert GOLD_STANDARD_INDEX.exists()
assert MARKER_SEED_INDEX.exists()
CHRONOSTRAIN_DB_DIR.mkdir(exist_ok=True, parents=True)
BLAST_DB_DIR.mkdir(exist_ok=True, parents=True)
print("All necessary files accounted for.")
```
